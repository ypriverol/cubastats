```python
import plotly_express as px
import pandas as pd
```

#Â Evolution of scientific publications overtime for countries with more than 100 citables papers in 1996 and less than 20k citables papers in 2018. The size of the buble is the number of citations per document.  

```python
data_publications = pd.read_csv('../publications-low.csv')
px.scatter(data_publications, x="Citable documents", y="Citations", animation_frame="Year", animation_group="Country",
           size="Citations per document", color="Country", hover_name="Country", 
           log_x = True, 
           size_max=45, range_x=[50,20000], range_y=[1,60000])
```
